```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/kaggle/input/mobile-phone-price-prediction/mobile phone price prediction.csv')
```

```python
df.head()
```

```python
df['Price']=df['Price'].str.replace(',','').astype(float)
temp=df.groupby('company')['Price'].mean()
plt.figure(figsize=(20,10))
plt.plot(temp.index,temp.values)
```

```python
df.columns
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Ram']=df['Ram'].apply(lambda x: x[0:2])
df['Battery']=df['Battery'].str.rstrip('mAh Battery')
df['Display']=df['Display'].str.rstrip('inches')
df['Inbuilt_memory']=df['Inbuilt_memory'].str.rstrip('GB inbuilt')
df['fast_charging']=df['fast_charging'].str.rstrip('W Fast Charging')
```

```python
df.rename(columns={'Ram':'Ram(GB)','Battery':'Battery(mAh)','Display':'Display(inches)','Inbuilt_memory':'Inbuilt_memory(GB)','fast_charging':'fast_charging(W)'},inplace=True)
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
obj_cols=['No_of_sim','Camera','External_Memory','company','Screen_resolution','Processor','Processor_name']
for col in obj_cols:
    df[col]=le.fit_transform(df[col])
```

```python
df['Name']=le.fit_transform(df['Name'])
```

```python
df.reset_index(drop=True)
```

```python
float_cols=['Rating','Display(inches)']
for col in float_cols:
    df[col]=pd.to_numeric(df[col],errors='coerce')
```

```python
df.info()
```

```python
int_cols=['Ram(GB)','Battery(mAh)','Android_version','Inbuilt_memory(GB)','fast_charging(W)']
for col in int_cols:
    df[col]=pd.to_numeric(df[col],errors='coerce')
```

```python
df['Price']=pd.to_numeric(df['Price'],errors='coerce')
df.info()
```

```python
df.head()
```

```python
sns.lineplot(x='Ram(GB)',y='Price',data=df)
```

```python
from sklearn.model_selection import train_test_split
df.columns
```

```python
X=df[['Name', 'Rating', 'Spec_score', 'No_of_sim', 'Ram(GB)', 'Battery(mAh)',
       'Display(inches)', 'Camera', 'External_Memory', 'Android_version',
       'company', 'Inbuilt_memory(GB)', 'fast_charging(W)',
       'Screen_resolution', 'Processor', 'Processor_name']]
y=df.Price
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='mean')
X_train=imputer.fit_transform(X_train)
X_test=imputer.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(pred,y_test)
```

```python
x = np.arange(len(y_test)) 

plt.figure(figsize=(20, 10))
plt.plot(x, y_test, label="True Values")
plt.plot(x, pred, label="Predicted Values")
plt.legend()
plt.show()
```
